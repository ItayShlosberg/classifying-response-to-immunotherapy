```python
def create_new_lines(sentence):
    parts_sentence = sentence.split(' ')
    div = 0
    new_sentence = ''
    for idx in range(len(parts_sentence)-1):
        div += 1
        new_sentence += parts_sentence[idx]+' '
        if div%6==0:
            new_sentence+= '\n'

    new_sentence = new_sentence + parts_sentence[-1]
    return new_sentence
```

```python
PATHWAYS_PATH = r'C:\Users\KerenYlab\Desktop\Technion studies\Keren laboratory\milestones\milestone 16 - 26.4.22\GEP_cohort_5.5.22\K21\pathways\MSigDB\GEPS_pathways.xlsx'
```

```python
main_pathways_df = pd.DataFrame(columns=['Program', 'Pathway', '# Genes in Overlap (k)', 'k/K', 'p-value', 'FDR q-value'])
for p_num in range(1, 22):
    if p_num==5:
        continue
    p_name = f'GEP{p_num}'
    curr_df = pd.read_excel(PATHWAYS_PATH, sheet_name=p_name)
    curr_df['Pathway'] = curr_df.Description.apply(lambda x: create_new_lines(x))
    curr_df['Program'] = p_num
    main_pathways_df = main_pathways_df.append(curr_df[main_pathways_df.columns])
```

```python
main_pathways_df = main_pathways_df.rename(columns={'# Genes in Overlap (k)': '# Genes in\nOverlap (k)'})
main_pathways_df
```
